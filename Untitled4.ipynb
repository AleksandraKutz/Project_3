```python
import pandas as pd
```

```python
Path = 'kids_per_code.csv'
```

```python
df_kids = pd.read_csv(Path)
```

```python
df_kids
```

```python
list(df_kids)
```

```python
columns_to_drop = [
    'Population Under 5 years',
    'Population 5 to 9 years',
    'Population 10 to 14 years',
    'Population 15 to 19 years',
    'Percent Population Under 5 years',
    'Percent Population 5 to 9 years',
    'Percent Population 10 to 14 years',
    'Male Population Under 5 years',
    'Male Population 5 to 9 years',
    'Male Population 10 to 14 years',
    'Female Population Under 5 years',
    'Female Population 5 to 9 years',
    'Female Population 10 to 14 years',
]
df_kids= df_kids.drop(columns=columns_to_drop)
```

```python
df_kids
```

```python
df_kids['Male Population Under 18 years (Selected)'] = df_kids['Population Under 18 years (Selected)'] - df_kids['Female Population Under 18 years (Selected)']
df_kids
```

```python
df_kids.rename(columns = {
    'Population Under 18 years (Selected)' : 'Population Under 18 years',
    'Percent Population Under 18 years (Selected)' : 'Percent Population Under 18 years',
    'Female Population Under 18 years (Selected)' : 'Female Population Under 18 years',
    'Male Population Under 18 years (Selected)' : 'Male Population Under 18 years'
}, inplace = True)
df_kids.info()
```

```python
df_kids['ZIP Code'] = df_kids['ZIP Code'].astype('string')
```

```python
df_kids.info()
```

```python
Path2 = 'pop_density_per_zip.csv'
```

```python
df_populat = pd.read_csv(Path2)
```

```python
df_populat
```

```python
df_populat = df_populat[['Zipcode', 'Population Density per sq mile']]
df_populat
```

```python
df_populat.rename(columns = {"Zipcode" : "ZIP Code"}, inplace = True)
```

```python
df_populat
```

```python
df_populat.info()
```

```python
df_populat['ZIP Code'] = df_populat['ZIP Code'].astype('string')
df_populat.info()
```

```python
kids_popul = pd.merge(df_kids, df_populat, on='ZIP Code', how='outer')
kids_popul
```

```python
kids_popul.to_csv("Population_table_for_sql.csv", index = False)
```
