```python
import pandas as pd
from pathlib import Path
```

```python
zipcode_area_path = Path('zipcode_area_size.csv')

zipcode_area_df = pd.read_csv(zipcode_area_path)

zipcode_area_df.head(5)
```

```python

zipcode_area_df = zipcode_area_df.rename(
    columns = {
        "Geographic Area Name (NAME)": "Zipcode",
        "Area (Land, in square meters) (AREALAND)": "Land Area sq meters",
        "Area (Water, in square meters) (AREAWATR)": "Water Area sq meters",
        "Area (Land, in square miles) (AREALAND_SQMI)": "Land Area sq miles",
        "Area (Water, in square miles) (AREAWATR_SQMI)": "Water Area sq miles",
        "Internal Point (Latitude) (INTPTLAT)": "Latitude",
        "Internal Point (Longitude) (INTPTLON)": "Longitude"
    }
)
zipcode_area_df.dtypes
```

```python

zipcode_area_df['Zipcode'] = zipcode_area_df['Zipcode'].str[-5:].astype(int)

zipcode_area_df.head(5)
```

```python
# zipcode_area_df['Land Area sq miles'] = pd.to_numeric(zipcode_area_df['Land Area sq miles'])

zipcode_area_df['Land Area sq miles']=  zipcode_area_df['Land Area sq miles'].str.replace(',','')

zipcode_area_df['Land Area sq miles'] = pd.to_numeric(zipcode_area_df['Land Area sq miles'])

zipcode_area_df.dtypes
```

```python
poverty_rate_path = Path('poverty_rate_per_zip.csv')

poverty_rate_df = pd.read_csv(poverty_rate_path)

poverty_rate_df.head(5)

poverty_rate_df.dtypes
```

```python
column_list = ['Zipcode', 'Land Area sq meters', 'Water Area sq meters', 'Land Area sq miles', 'Water Area sq miles', 'Latitude', 'Longitude', 'Population']

pop_density_df = zipcode_area_df.merge(poverty_rate_df, left_on='Zipcode', right_on='Zipcode')[column_list]

pop_density_df.head(5)
```

```python
pop_density_df['Population Density per sq mile'] = pop_density_df['Population']/pop_density_df['Land Area sq miles']

pop_density_df.tail(5)
```

```python
pop_density_df.to_csv("pop_density_per_zip.csv", index=False)
```
